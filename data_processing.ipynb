```python
cleanedDataPcf = cleanedDataPcf[cleanedDataPcf["product_title"].str.contains("pacifier") | 
                                cleanedDataPcf["product_title"].str.contains("Pacifier")]

cleanedDataMcw = cleanedDataMcw[cleanedDataMcw["product_title"].str.contains("microwave") | 
                                cleanedDataMcw["product_title"].str.contains("Microwave")]

cleanedDataHdr = cleanedDataHdr[cleanedDataHdr["product_title"].str.contains("hair dryer")  |
                cleanedDataHdr["product_title"].str.contains("Hair dryer") |
                cleanedDataHdr["product_title"].str.contains("Hair Dryer") |
              cleanedDataHdr["product_title"].str.contains("hair Dryer")]
```

```python
cleanedDataPcf.to_csv("CleanedData/pacifier.csv")

cleanedDataMcw.to_csv("CleanedData/microwave.csv")

cleanedDataHdr.to_csv("CleanedData/hair_dryer.csv")
```
